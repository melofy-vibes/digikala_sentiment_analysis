```python
from datasets import load_dataset
import random
import pandas as pd
```

لینک دیتاست

https://huggingface.co/datasets/RadeAI/Digikala_comments_products

```python
# 1️⃣ load کامل
ds = load_dataset(
    "RadeAI/Digikala_comments_products",
    data_files="digikala-comments.csv",
    split="train"
)

# 2️⃣ shuffle
ds = ds.shuffle(seed=42)

negatives = []
positives = []

for sample in ds:
    rate = sample.get("rate")
    text = sample.get("body")

    if not text or rate is None:
        continue

    if rate in [1, 2] and len(negatives) < 75000:
        negatives.append({
        "product_id": sample.get("product_id"),
        "body": text,
        "rate": rate
        })

    elif rate in [4, 5] and len(positives) < 75000:
        positives.append({
        "product_id": sample.get("product_id"),
        "body": text,
        "rate": rate
        })

    if len(negatives) >= 75000 and len(positives) >= 75000:
        break

print("Negative:", len(negatives))
print("Positive:", len(positives))
```

```python
neg_df = pd.DataFrame(negatives)
pos_df = pd.DataFrame(positives)

neg_df["label"] = 0
pos_df["label"] = 1

final_df = pd.concat([neg_df, pos_df]).sample(frac=1, random_state=42)

final_df.to_csv(
    r"data/digikala_150k.csv",
    index=False,
    encoding="utf-8-sig"
)

print("✅ Done!")
```
